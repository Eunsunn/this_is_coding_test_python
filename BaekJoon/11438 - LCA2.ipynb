```python
# https://www.acmicpc.net/problem/11438

import sys
from collections import deque
from math import ceil, log2

# O(N + MlogN)
def solution():
    input = sys.stdin.readline
    n = int(input().rstrip()) # 노드 개수
    root = 1
    connected = [[] for _ in range(n + 1)]
    for _ in range(n - 1):
        node1, node2 = map(int, input().split())
        connected[node1].append(node2)
        connected[node2].append(node1)
    
    # 조상 테이블 - O(N)
    parent = [0 for _ in range(n + 1)]
    child = [[] for _ in range(n + 1)]
    visited = [False] * (n + 1)
    level = [0 for _ in range(n + 1)]
    level[root] = 1
    max_level = 17 # log2(10 ** 5  + 1) <= 17
    # ancestor[i][j] = i의 2 ** j번째 조상
    ancestor = [[0] * max_level for _ in range(n + 1)]
    q = deque()
    q.append(root)
    while q:
        cur = q.popleft()
        visited[cur] = True
        iterate = ceil(log2(level[cur]))
        for exponent in range(iterate):
            if exponent == 0:
                ancestor[cur][exponent] = parent[cur]
            else:
                half_ancestor = ancestor[cur][exponent - 1]
                ancestor[cur][exponent] = ancestor[half_ancestor][exponent - 1]
        
        for next_node in connected[cur]:
            if visited[next_node]:
                continue
            level[next_node] = level[cur] + 1
            child[cur].append(next_node)
            parent[next_node] = cur
            q.append(next_node)

    # 공통 조상 찾기
    m = int(input().rstrip())
    for _ in range(m):
        node1, node2 = map(int, input().split())
        # 두 노드의 레벨을 맞추고 - O(logN)
        # node2이 크거나 같은 상태로 초기화
        if level[node1] > level[node2]:
            node1, node2 = node2, node1
        while level[node1] < level[node2]:
            diff = level[node2] - level[node1]
            diff = int(log2(diff))
            node2 = ancestor[node2][diff]
        # 레벨만큼 부모로 올라간다 - O(logN)
        while node1 != node2:
            # 같지 않은 최대 부모까지 올라간다
            iterate = ceil(log2(level[node1]))
            exponent = 0
            for i in reversed(range(iterate)):
                if ancestor[node1][i] != ancestor[node2][i]:
                    exponent = i
                    break
            node1, node2 = ancestor[node1][exponent], ancestor[node2][exponent]
        print(node1)


# def solution2():
#     import sys
#     input = sys.stdin.readline
#     sys.setrecursionlimit(int(1e5))
#     LOG = 21 # 2*20 = 1,000,000

#     n = int(input().rstrip())
#     parent = [[0] * LOG for _ in range(n + 1)] # 부모 노드 정보
#     d = [0] * (n + 1) # 각 노드까지 깊이
#     c = [0] * (n + 1) # 각 노드까지 깊이가 계산되었는지 여부
#     graph = [[] for _ in range(n + 1)] # 그래프 정보

#     for _ in range(n - 1):
#         a, b = map(int, input().split())
#         graph[a].append(b)
#         graph[b].append(a)

#     # 루트 노드부터 시작해 깊이를 구하는 함수
#     def dfs(x, depth):
#         c[x] = True
#         d[x] = depth
#         for y in graph[x]:
#             if c[y]: # 이미 방문한 경우 패스
#                 continue
#             parent[y][0] = x # 부모저장
#             dfs(y, depth + 1)

#     # 전체 부모관계를 설정하는 함수
#     def set_parent():
#         dfs(1, 0)
#         for i in range(1, LOG):
#             for j in range(1, n + 1):
#                 # 2 ** j = 2 * (2 ** (j - 1))
#                 parent[i][j] = parent[parent[i][j - 1]][j - 1]
    
#     # A와 B의 최소 공통조상을 찾는 함수
#     def lca(a, b):
#         # b가 더 깊도록 설정
#         if d[a] > d[b]:
#             a, b = b, ancestor
#         # 깊이를 같게 맞춰주기
#         for i in range(LOG, -1, -1):
#             # 최대한 뛰어넘을 수 있는 차이만큼 뛰어넘기
#             if d[b] - d[a] >= (1 << i): # 1 << i : 2의 지수
#                 b = parent[b][i]
#         # base case: 같으면 종료
#         if a == b:
#             return a
#         for i in range(LOG, -1, -1):
#             # 다르면 부모를 타고 올라간다
#             # 부모를 타고가면서 인덱스(i)가 유지된다 (그 이상 볼 필요 x)
#             if parent[a][i] != parent[b][i]:
#                 a = parent[a][i]
#                 b = parent[b][i]
#         return parent[a][0]

#     set_parent()
#     m = int(input().rstrip())
#     for i in range(m):
#         a, b = map(int, input().split())
#         print(lca(a, b))

        
solution()
```
