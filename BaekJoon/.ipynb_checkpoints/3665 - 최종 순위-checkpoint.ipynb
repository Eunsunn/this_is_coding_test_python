				<div id="problem_description" class="problem-text">
				<p>올해 ACM-ICPC 대전 인터넷 예선에는 총 n개의 팀이 참가했다. 팀은 1번부터 n번까지 번호가 매겨져 있다. 놀랍게도 올해 참가하는 팀은 작년에 참가했던 팀과 동일하다.</p>

<p>올해는 인터넷 예선 본부에서는 최종 순위를 발표하지 않기로 했다. 그 대신에 작년에 비해서 상대적인 순위가 바뀐 팀의 목록만 발표하려고 한다. (작년에는 순위를 발표했다) 예를 들어, 작년에 팀 13이 팀 6 보다 순위가 높았는데, 올해 팀 6이 팀 13보다 순위가 높다면, (6, 13)을 발표할 것이다.</p>

<p>창영이는 이 정보만을 가지고 올해 최종 순위를 만들어보려고 한다. 작년 순위와 상대적인 순위가 바뀐 모든 팀의 목록이 주어졌을 때, 올해 순위를 만드는 프로그램을 작성하시오. 하지만, 본부에서 발표한 정보를 가지고 확실한&nbsp;올해 순위를 만들 수 없는 경우가 있을 수도 있고, 일관성이 없는 잘못된 정보일 수도 있다. 이 두 경우도 모두 찾아내야 한다.</p>

	<li>두 정수 a<sub>i</sub>와 b<sub>i</sub>를 포함하고 있는 m줄. (1 ≤ a<sub>i</sub> &lt; b<sub>i</sub> ≤ n) 상대적인 등수가 바뀐 두 팀이 주어진다. 같은 쌍이 여러 번 발표되는 경우는 없다.</li>

# 은선

```python
# https://www.acmicpc.net/problem/3665
# 얘랑 문제가 비슷하다: https://programmers.co.kr/learn/courses/30/lessons/49191?language=python3
import sys
from collections import deque

def solution():
    input = sys.stdin.readline
    cases = int(input())
    for _ in range(cases):
        n = int(input())
        last_year = list(map(int, input().split()))
        last_year = last_year[::-1] # 순위가 낮은 팀부터
        # 작년 경기 결과
        indegree = [0] * (n + 1)
        graph = [[False] * (n + 1) for _ in range(n + 1)]
        for i in range(n):
            loser = last_year[i]
            indegree[loser] = i
            for j in range(i+1, n):
                winner = last_year[j]
                graph[loser][winner] = True # loser -> winner
            
        
        # 바뀐 경기 결과
        m = int(input())
        for _ in range(m):
            team1, team2 = map(int, input().split())
            loser, winner = None, None
            # team1(loser) -> team2(winner)
            if graph[team1][team2]:
                loser, winner = team1, team2
            # team2(loser) -> team1(winner)
            elif graph[team2][team1]:
                loser, winner = team2, team1
            # 우선순위 바꾸기
            graph[loser][winner] = False
            graph[winner][loser] = True
            indegree[loser] += 1
            indegree[winner] -= 1


        # 결과 출력: 위상 정렬
        def tropology_sort():
            q = deque()

            # 내차수가 0인 노드로 초기화
            for i in range(1, n + 1):
                if indegree[i] == 0:
                    q.append(i)
            
            # 큐가 빌 때까지 위상정렬
            result = []
            while q:
                # 순위가 같은 경우 : 확실한 순위 x
                if len(q) > 1:
                    return "?"
                now = q.popleft()
                result.append(now)
                # 현재 노드와 연결된 노드에대해
                for next in range(1, n + 1):
                    # 차수를 감소시키고
                    if graph[now][next]:
                        indegree[next] -= 1
                        # 새롭게 차수가 0이 되는 노드를 큐에 삽입
                        if indegree[next] == 0:
                            q.append(next)
            
            # 모두 방문 불가: 데이터의 일관성 x
            if len(result) < n:
                return "IMPOSSIBLE"
            else:
            # 순위가 높은 순 출력
                result = [str(i) for i in result[::-1]]
                return ' '.join(result) 
        
        print(tropology_sort())

if __name__ == "__main__":
    solution()
```
