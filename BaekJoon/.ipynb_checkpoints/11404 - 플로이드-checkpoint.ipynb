# 플로이드
<h2>문제</h2>
</div>
<div id="problem_description" class="problem-text">
<p>n(2&nbsp;≤ n ≤ 100)개의 도시가 있다. 그리고 한 도시에서 출발하여 다른 도시에 도착하는 m(1 ≤ m ≤ 100,000)개의 버스가 있다. 각 버스는 한 번 사용할 때 필요한 비용이 있다.</p>

<p>모든 도시의 쌍 (A, B)에 대해서 도시 A에서 B로 가는데 필요한 비용의 최솟값을 구하는 프로그램을 작성하시오.</p>
                    
<section id="input" class="problem-section">

					</div>

					<p>첫째 줄에 도시의 개수 n이 주어지고 둘째 줄에는 버스의 개수 m이 주어진다. 그리고 셋째 줄부터 m+2줄까지 다음과 같은 버스의 정보가 주어진다. 먼저 처음에는 그 버스의 출발 도시의 번호가 주어진다. 버스의 정보는 버스의 시작 도시 a, 도착 도시 b, 한 번 타는데 필요한 비용 c로 이루어져 있다. 시작 도시와 도착 도시가 같은 경우는 없다. 비용은 100,000보다 작거나 같은 자연수이다.</p>

<p>시작 도시와 도착 도시를 연결하는 노선은 하나가 아닐 수 있다.</p>

# 은선

```python
# https://www.acmicpc.net/problem/11404
# 모든 노드에서 모든 노드로 가는 최단 거리
# -> 플로이드 워셜 알고리즘

import sys

def solution():
    input = sys.stdin.readline
    # 노드 개수, 간선 개수
    n = int(input())
    m = int(input())
    INF = int(1e5) * n
    dist = [[INF] * (n + 1) for _ in range(n + 1)]
    # 간선으로 거리 초기화
    for _ in range(m):
        start, end, cost = map(int, input().split())
        # 노선이 중복되는 경우 최소값 선택
        dist[start][end] = min(dist[start][end], cost)
    # 자기 자신으로 가는 거리는 0으로 초기화
    for i in range(1, n + 1):
        dist[i][i] = 0
        
    # idx번째 노드를 거쳐 가기
    def drop_by(idx):
        for i in range(1, n + 1):
            if i == idx: continue
            for j in range(1, n + 1):
                if j == i or j == idx: continue
                dist[i][j] = min(dist[i][j], dist[i][idx] + dist[idx][j])

    # 플로이드 워셜
    def floyd():
        # 노드를 거쳐가면서 거리 갱신
        for i in range(1, n + 1):
            drop_by(i)
    
    # 플로이드 워셜 수행
    floyd()
    # 결과 출력
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            if dist[i][j] == INF:
                print(0, end=" ")
            else:
                print(dist[i][j], end=" ")
        print()


if __name__ == "__main__":
    solution()
```
