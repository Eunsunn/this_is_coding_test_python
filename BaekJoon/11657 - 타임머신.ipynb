# 은선

```python
# 벨만 포드 알고리즘
# - 음수 간선이 있는 경우의 최단거리
# https://www.acmicpc.net/problem/11657

import sys

    
def solution():
    input = sys.stdin.readline
    INF = int(1e9)
    # 노드 개수, 간선 개수
    n, m = map(int, input().split())
    # 간선 정보 리스트
    edges = []
    # 최단거리 테이블을 무한으로 초기화
    dist = [INF] * (n + 1)

    # 모든 간선 정보 입력받기
    for _ in range(m):
        a, b, c = map(int, input().split())
        edges.append((a, b, c))

    # 벨만포드 알고리즘
    def bf(start):
        # 시작 노드의 거리 초기화
        dist[start] = 0
        # 전체 n번의 라운드(=노드개수) 반복
        # 시작노드를 제외한 n - 1개의 노드 탐색 + 마지막 사이클 체크
        for i in range(n):
            # 매 반복마다 모든 간선을 확인
            for j in range(m):
                cur = edges[j][0]
                next_node = edges[j][1]
                cost = edges[j][2]
                # 현재 간선을 거쳐서 거리가 더 짧은 경우
                if dist[cur] != INF and dist[next_node] > dist[cur] + cost:
                    dist[next_node] = dist[cur] + cost
                    # n번째 라운드에서도 값이 갱신된다면 음수 순환 존재
                    if i == n - 1:
                        return True
        return False

    # 1번 노드에서 시작
    negative_cycle = bf(1)

    if negative_cycle:
        print("-1")
    else:
        # 1번 노드에서 출발해 다른 노드로 가는 최단거리 출력
        for i in range(2, n + 1):
            # 도달할 수 없는 경우 -1 출력
            if dist[i] == INF:
                print("-1")
            # 도달할 수 있는 경우 거리 출력
            else:
                print(dist[i])


solution()
```
