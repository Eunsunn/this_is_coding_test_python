# 은선

```python
# binary index tree
# https://www.acmicpc.net/problem/2042

import sys

def solution():
    input = sys.stdin.readline
    
    # 데이터 개수, 변경 횟수, 구간합 횟수
    n, m, k = map(int, input().split())
    
    # 전체 데이터 개수는 최대 1,000,000개
    arr = [0] * (n + 1) # 데이터 저장
    tree = [0] * (n + 1) # binary index tree의 값(합)을 저장

    # i번째까지 누적합을 계산하는 함수
    def prefix_sum(i):
        result = 0
        while i > 0:
            result += tree[i]
            # 0이 아닌 마지막 비트의 값만큼 빼면서 인덱스 이동
            i -= (i & -i)
        return result

    # i번째 수를 dif만큼 더하는 함수 (tree에서)
    def update(i, dif):
        while i <= n:
            tree[i] += dif
            # 0이 아닌 마지막 비트의 값만큼 인덱스를 더하면서 업데이트
            i += (i & -i)
        
    # start부터 end까지의 구간합을 구하는 함수
    def interval_sum(start, end):
        return prefix_sum(end) - prefix_sum(start - 1)

    for i in range(1, n + 1):
        x = int(input())
        arr[i] = x
        update(i, x)

    for _ in range(m + k):
        a, b, c = map(int, input().split())
        if a == 1:
            update(b, c - arr[b]) # 바뀐 크기(dif)만큼만 tree update 적용
            arr[b] = c
        elif a == 2:
            print(interval_sum(b, c))
            
solution()
```
