# 은선

```python
# https://www.acmicpc.net/problem/2003

def solution():
    # N개의 데이터, 합이 M이 되는 경우
    n, m = map(int, input().split())
    # 데이터 리스트
    number = list(map(int, input().split()))

    tmp_sum = 0
    end = 0
    cnt = 0

    for start in range(n):
        # 합이 작은경우 end를 이동시키고
        while tmp_sum < m and end < n:
            tmp_sum += number[end]
            end += 1
        if tmp_sum == m:
            cnt += 1
        tmp_sum -= number[start]
    return cnt


if __name__ == "__main__":
    print(solution())
```
