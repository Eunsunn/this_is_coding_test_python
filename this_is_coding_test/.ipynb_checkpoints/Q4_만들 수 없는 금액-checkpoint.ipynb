# 만들 수 없는 금액
<p> 
    N개의 동전을 이용하여 만들 수 없는 양의 정수 금액 중 최솟값을 구하는 프로그램을 작성하시오.
    입력조건: 첫째줄에는 동전의 개수를 나타내는 양의 정수 N (1 <= N <=1000)
                  둘째줄에는 동전의 화폐 단위를 나타내는 N개의 자연수가 주어지며, 각 자연수는 공백으로 구분한다. (각 화폐의 단위는 1,000,000 이하의 자연수)
    출력조건: 첫째줄에 주어진 동전들로 만들 수 없는 양의 정수 금액 중 최솟값
</p>

    입력예시1:
    3
    3 5 7
    출력예시1:
    1
    
    입력예시2:
    5
    3 2 1 1 9
    출력예시2:
    8


# 은선

```python
def solution():
    # 직접 입력받을 때
    # _ = int(input())
    # money = map(int, input().split())

    # test case
    # money = [3, 2, 1, 1, 9] # 정답 8
    # money = [5] # 정답 1
    # money = [1, 1, 1] # 정답 4
    money = [1, 2, 3, 4] # 정답 11

    # "최소" 금액 : greedy, 정렬
    money.sort()

    # prev: 이전에 만든 것중 최대 금액
    # 1~prev까지 모두 만들 수 있다
    prev = 0 
    for m in money:
        # 실패: m을 그대로 넣어도 실패
        if prev + 1 < m: return prev + 1
        # 성공: 이전 최대 금액에서 m만큼 더 만들 수 있다
        prev += m
    
    # 중간에 실패하지 않는 경우는 최대금액 + 1
    return prev + 1


if __name__ == "__main__":
    print(solution())
```
