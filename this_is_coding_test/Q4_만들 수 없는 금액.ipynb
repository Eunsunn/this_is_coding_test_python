```python
def solution():
    #   
    # _ = int(input())
    # money = map(int, input().split())

    # test case
    # money = [3, 2, 1, 1, 9] #  8
    # money = [5] #  1
    # money = [1, 1, 1] #  4
    money = [1, 2, 3, 4] #  11

    # ""  : greedy, 
    money.sort()

    # prev:     
    # 1~prev    
    prev = 0 
    for m in money:
        # : m   
        if prev + 1 < m: return prev + 1
        # :    m    
        prev += m
    
    #      + 1
    return prev + 1


if __name__ == "__main__":
    print(solution())
```
