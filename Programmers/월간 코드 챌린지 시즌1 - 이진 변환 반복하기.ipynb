<h6 class="guide-section-title">문제 설명</h6>
<div class="markdown solarized-dark"><p>0과 1로 이루어진 어떤 문자열 x에 대한 이진 변환을 다음과 같이 정의합니다.</p>

<ol>
<li>x의 모든 0을 제거합니다.</li>
<li>x의 길이를 c라고 하면, x를 <q>c를 2진법으로 표현한 문자열</q>로 바꿉니다.</li>
</ol>

<p>예를 들어, <code>x = "0111010"</code>이라면, x에 이진 변환을 가하면 <code>x = "0111010" -&gt; "1111" -&gt; "100"</code> 이 됩니다.</p>

<p>0과 1로 이루어진 문자열 s가 매개변수로 주어집니다. s가 <q>1</q>이 될 때까지 계속해서 s에 이진 변환을 가했을 때, 이진 변환의 횟수와 변환 과정에서 제거된 모든 0의 개수를 각각 배열에 담아 return 하도록 solution 함수를 완성해주세요.</p>

<hr>

<h5>제한사항</h5>

<ul>
<li>s의 길이는 1 이상 150,000 이하입니다.</li>
<li>s에는 '1'이 최소 하나 이상 포함되어 있습니다.</li>
</ul>

# 은선
- 문제: https://programmers.co.kr/learn/courses/30/lessons/70129

```python
def solution(s):
    remove_zeros = 0
    cnt_convert = 0
    number = s
    
    while number != ['1']:
        cnt_convert += 1
        # 0을 제거
        num_zeros = 0
        for i in number:
            num_zeros += (i=="0")
        remove_zeros += num_zeros
        
        # 길이 계산
        quotient = len(number) - num_zeros
        
        # 길이 2진수로
        next_number = []
        while quotient > 1:
            remain = quotient % 2
            next_number.insert(0, str(remain))
            quotient = int(quotient/2)
        next_number.insert(0, str(quotient))
        number = next_number


    return [cnt_convert, remove_zeros]
```
