# 은선

```python
import sys

def solution(begin, target, words):
    if not target in words:
        return 0
    word_length = len(begin)
    num_word = len(words)

    # 단어간 연결 여부 이중 리스트 만들기
    def adjacent(words):
        connected = [[False for _ in range(num_word)] for _ in range(num_word)]
        for i in range(num_word):
            word = words[i]
            for j in range(i+1, num_word):
                other = words[j]
                same_char = [True if word[k]==other[k] else False for k in range(word_length)]
                if sum(same_char)==word_length-1:
                    connected[i][j] = connected[j][i] = True
        return connected

    # begin 단어와 연결된 단어 리스트 만들기
    def begin_adjacent(begin, start):
        connected = []
        for idx in range(num_word):
            word = words[idx]
            same_char = [True if begin[i]==word[i] else False for i in range(word_length)]
            if sum(same_char)==word_length-1:
                connected.append(idx)
        return connected

    # DFS로 단어 연결 최소 시간 찾기
    def dfs(start_idx, target_idx, t, min_t, connected, visited):
        # Base case
        if start_idx == target_idx:
            return t
        if t >= min_t:
            return -1
        
        # 다음 노드 재귀 탐색
        for i in range(num_word):
            if connected[start_idx][i] and not visited[i]:
                visited[i] = True
                ret = dfs(i, target_idx, t+1, min_t, connected, visited)
                if ret > 0:
                    min_t = ret
                visited[i] = False
        
        return min_t


    min_t = sys.maxsize
    target = words.index(target)
    connected = adjacent(words)
    for start in begin_adjacent(begin, words):
        visited = [False for _ in range(num_word)]
        visited[start] = True
        ret = dfs(start, target, 1, min_t, connected, visited)
        if ret > 0:
            min_t = ret

    answer = 0
    if min_t < sys.maxsize:
        answer = min_t
    return answer
```
