      <div class="markdown solarized-dark"><p>레스토랑을 운영하고 있는 <strong><q>스카피</q></strong>는 레스토랑 내부가 너무 낡아 친구들과 함께 직접 리모델링 하기로 했습니다. 레스토랑이 있는 곳은 스노우타운으로 매우 추운 지역이어서 내부 공사를 하는 도중에 주기적으로 외벽의 상태를 점검해야 할 필요가 있습니다.</p>

# 은선

```python
from collections import defaultdict

def solution(n, weak, dist):
    W, D = len(weak), len(dist)
    # 일했는지 - dist에서 인덱스
    worked = [False] * D
    # 찾았는지 - 거리
    find = [False] * n
    # 다음 취약지점 위치
    next_weak = dict()
    for i in range(len(weak)):
        next_i = (i + 1) % len(weak)
        next_weak[weak[i]] = weak[next_i]
    wf_cnt = [0, 0] # worked, find cnt


    def move_dist(loc, dist_idx):
        wf_cnt[0] += 1 # worked ++
        wf_cnt[1] += 1 # find ++
        worked[dist_idx] = True
        find[loc] = True
        checked_ls = [loc]
        move, tmp_loc = 0, loc
        while True:
            if tmp_loc < next_weak[tmp_loc]: 
                diff = next_weak[tmp_loc] - tmp_loc
            else:
                diff = (n - tmp_loc) + next_weak[tmp_loc]
            if dist[dist_idx] < move + diff: break
            tmp_loc = next_weak[tmp_loc]
            if find[tmp_loc]: break
            wf_cnt[1] += 1 # find ++
            find[tmp_loc] = True
            checked_ls.append(tmp_loc)
            move += diff

        # 다음 weak 위치
        next_loc = next_weak[checked_ls[-1]]
        return next_loc, checked_ls


    def uncheck(dist_idx, checked_ls):
        wf_cnt[0] -= 1 # worked --
        worked[dist_idx] = False
        for element in checked_ls:
            wf_cnt[1] -= 1 # find --
            find[element] = False


    def check_wall(loc, cur_dist):
        # base case. 더이상 탐색 불가
        if wf_cnt[0] == D:
            return -1

        # loc에서 cur_dist만큼 이동
        next_loc, checked_ls = move_dist(loc, cur_dist)

        # base case. 탐색 완료
        num_work = wf_cnt[0]
        if wf_cnt[1] == W:
            uncheck(cur_dist, checked_ls)
            return num_work
        # 조기종료
        if answer > 0 and num_work >= answer:
            uncheck(cur_dist, checked_ls)
            return -1

        # 다음 탐색
        ret = -1
        for next_dist_idx in range(len(dist)):
            if worked[next_dist_idx]: continue
            tmp = check_wall(next_loc, next_dist_idx)
            if tmp > 0:
                if ret == -1:
                    ret = tmp
                else:
                    ret = min(ret, tmp)
        
        # 현재 이동 풀어주기
        uncheck(cur_dist, checked_ls)
        
        return ret


    answer = -1
    # 모든 weak 시작점에서
    for wall_start in weak:
        # 모든 일꾼의 순서로 탐색한다
        for dist_idx in range(len(dist)):
            # 탐색
            tmp = check_wall(wall_start, dist_idx)
            # 이번 탐색값이 양수면 answer = min(answer, tmp)로 갱신
            if tmp > 0:
                if answer == -1:
                    answer = tmp
                else:
                    answer = min(answer, tmp)
    
    return answer
```
